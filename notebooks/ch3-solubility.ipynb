# अध्याय 3: विलेयता का पूर्वानुमान

ताप बनाम विलेयता के डेटा से किसी मध्यवर्ती ताप पर विलेयता का पूर्वानुमान।

_विज्ञान में कृत्रिम बुद्धिमत्ता_ — companion notebook. ▶ Runtime → Run all चलाएँ।

```python
import numpy as np

T = np.array([10, 20, 30, 40, 50])
S = np.array([31, 34, 37, 40, 43])

a, b = np.polyfit(T, S, 1)
T_new = 35
print("predicted solubility at 35 C =", round(a*T_new + b, 1), "g/100mL")
```
