# अध्याय 4: DNA क्षारक-गणना एवं GC content

किसी DNA अनुक्रम में A,T,G,C की गिनती तथा GC प्रतिशत — Bioinformatics का पहला कदम।

_विज्ञान में कृत्रिम बुद्धिमत्ता_ — companion notebook. ▶ Runtime → Run all चलाएँ।

```python
dna = "ATGCGGATTACGCTAGCATGGC"

for base in "ATGC":
    print(base, "count =", dna.count(base))

gc = (dna.count("G") + dna.count("C")) / len(dna) * 100
print("GC content =", round(gc, 1), "%")
```
