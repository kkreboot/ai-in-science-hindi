```python
import numpy as np

# experiment data: force F (newton) vs extension x (metre)
x = np.array([0.02, 0.04, 0.06, 0.08, 0.10])
F = np.array([1.0, 2.1, 2.9, 4.2, 5.0])

m, c = np.polyfit(x, F, 1)
print("spring constant k =", round(m, 2), "N/m")
```
