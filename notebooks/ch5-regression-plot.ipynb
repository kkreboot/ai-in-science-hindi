# अध्याय 5: डेटा का ग्राफ़ एवं regression

डेटा पर सर्वोत्तम रेखा निकालकर scatter-plot के साथ दर्शाना।

_विज्ञान में कृत्रिम बुद्धिमत्ता_ — companion notebook. ▶ Runtime → Run all चलाएँ।

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0.02, 0.04, 0.06, 0.08, 0.10])
y = np.array([1.0, 2.1, 2.9, 4.2, 5.0])

m, c = np.polyfit(x, y, 1)
print("slope m =", round(m, 2), " intercept c =", round(c, 2))

plt.scatter(x, y, label="data")
plt.plot(x, m*x + c, color="red", label="best-fit line")
plt.xlabel("x"); plt.ylabel("y"); plt.legend()
plt.show()
```
